## Convert Age into Buckets

```python
import pandas as pd
```

```python
PATH_TO_CSV = "../data/raw/raw_cleaned/final_test.csv"
```

```python
df = pd.read_csv(PATH_TO_CSV)
```

```python
df.head()
```

```python
# Age Bucket to apply
def age_bucket(age):
    if age < 0:
        return 'Invalid'
    elif age < 13:
        return 'Child'
    elif age < 18:
        return 'Teen'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'
```

```python
df['age'] = df['age'].apply(age_bucket)
```

```python
df.head()
```

```python
df['age'].value_counts()
```

```python
df.to_csv(PATH_TO_CSV)
```
