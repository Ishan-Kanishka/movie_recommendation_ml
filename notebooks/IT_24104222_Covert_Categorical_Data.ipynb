## Convert Categorical Data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
PATH_TO_CSV = "../data/raw/raw_cleaned/final_test.csv"
```

```python
df = pd.read_csv(PATH_TO_CSV)
```

```python
df.info()
```

```python
df.head(2)
```

### Gender

```python
lb_encoder = LabelEncoder()
```

```python
df['gender'] = lb_encoder.fit_transform(df['gender'])
```

```python
df.head()
```

## Occupation

```python
list(df['occupation'].values)
```

```python
df['occupation'] = lb_encoder.fit_transform(df['occupation'])
```

```python
df.head()
```

```python
df['occupation'].value_counts()
```

## Age

```python
# age_categories = df['age'].value_counts().index.tolist()
age_categories = ['Child', 'Teen', 'Adult', 'Senior']
print("Categories by frequency:", age_categories)
```

```python
ordinal_encoder = OrdinalEncoder(categories=[age_categories])
```

```python
df.head()
```

```python
df['age'].value_counts()
```

```python
df['age'] = ordinal_encoder.fit_transform(df[['age']])
```

```python
df.head()
```

```python
df['age'].value_counts()
```

```python
df.head()
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df.head()
```

```python
df.to_csv(PATH_TO_CSV, index=False)
```
